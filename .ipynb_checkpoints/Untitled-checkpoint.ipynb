## Cleaning text
## Dealing with review column

```python
df[['review']]
```

```python
# Filtering all the airlines which are Not Verified by the passenger:-

df.loc[df['review'].str.contains('Not Verified')]
```

```python
# Filtering all the airlines which are Trip Verified by the passenger:-

df.loc[df['review'].str.contains('Trip Verified')]
```

```python
# Creating new column named : Check Verify and storing whether the airline is verified or not by passenger:-

def fungame(value):
    if 'Trip Verified' in value:
        return 'Verified'
    elif 'Not Verified' in value:
        return 'Not Verified'
    else:
        return None
```

```python
df['check verify'] = df['review'].apply(fungame)
```

```python
# New column named: Check Verify store three types of values according to the review:-

df['check verify'].unique()
```

```python
df.head()
```

```python
df['review'] = df['review'].str.replace("? Trip Verified |", "")
df['review'] = df['review'].str.replace("Not Verified |", "")
df['review'] = df['review'].str.replace(" | ", "")
df['review'] = df['review'].str.replace(r'\b(\w{1,3})\b', "")
```
