## â‘¢ Dealing with route column

### â¡ Create a new column named: 'start' and store the value from where the journey start

```python
def journeystart(val):
    if pd.isna(val):
        return None
    elif " to " in val:
        return val.split(" to ")[0]
    else:
        return None

df['start'] = df['route'].apply(journeystart)
```

### â¡ Create a new column named: 'via' and store the value of via journey travel

```python
def journeyvia(val):
    if pd.isna(val):
        return None
    elif "via" in val:
        mid = val.split("via")[-1]
        return mid
    else:
        return None    
```

```python
df['via'] = df['route'].apply(journeyvia)
```

### â¡ Create a new column named: 'end' and store the value where destination arrived

```python
def journeyend(val):
    if pd.isna(val):
        return None
    elif " to " in val and " via " not in val:
        value = val.split(" to ")[-1]
        return value
    elif " to " in val and " via " in val:
        values = val.split(" to ")[-1]
        value = values.split(" via ")[0]
        return value
    else: 
        return None 
```

```python
df['end'] = df['route'].apply(journeyend)
```

## â‘¤ Airline - Aircraft

The goal of this below  code is to standardize and complete a dataset containing airline, aircraft related ratings. 
It does so by ensuring that the rating columns have a consistent numeric data type and then calculates average ratings for specific categories (like seat comfort or food quality) within partitions determined by the airline and aircraft. Any missing ratings are replaced with the corresponding average for that specific airline and aircraft combination. This process ensures that the dataset is uniform, with no missing values in the rating columns, making it more suitable for analysis and comparisons of airline services across different categories and carriers.

```python
rating_columns = ['seat_comfort', 'cabin_staff_service', 'food_and_beverages', 'ground_service', 'value_for_money', 'inflight_entertainment', 'wifi_and_connectivity']

for col_name in rating_columns:
    df[col_name] = df[col_name].astype('double')

# Calculate the average for each rating category within the specified partition
partition_columns = ["airline", "aircraft"]
for col_name in rating_columns:
    df[col_name + '_avg'] = df.groupby(partition_columns)[col_name].transform('mean')

# Replace null values in rating columns with the corresponding average
for col_name in rating_columns:
    df[col_name] = df[col_name].fillna(df[col_name + '_avg'])

# Drop the average columns
average_columns = [col_name + '_avg' for col_name in rating_columns]
df = df.drop(average_columns, axis=1)
df.head()
```

```python
from collections import Counter

# Tokenize the words
word_tokens = []
for text in df['corpus']:
    for word in text.split():
        word_tokens.append(word)

# Calculate word frequencies
word_freq = Counter(word_tokens)

# Display the word frequencies
for word, freq in word_freq.items():
    print(f"{word}: {freq}")
```

###  TextBlob module
â€¢ TextBlob is a versatile library for various natural language processing (NLP) tasks, its built-in sentiment analysis functionality is one of its notable features. Sentiment analysis is indeed one of the main purposes for which TextBlob is commonly use

```python
# Word with its count and sentiment score.

from textblob import TextBlob

print("{:<15}{:<15}{:<10}".format("word", 'count', 'sentiment_score'))
for word, freq in word_freq.items():
    specific_word = word
    blob = TextBlob(specific_word)
    
    # Get the sentiment polarity, which is a value between -1 and 1.
    sentiment_score = blob.sentiment.polarity
    print("{:<15}{:<15}{:<10}".format(word, freq, sentiment_score))
```

###  â€¢ Create a new DataFrame named 'df1' to store words, their counts, and their sentiment scores. This will help extract insight about the performance of a list of airlines that are performing well.

```python
import pandas as pd

data = []
from textblob import TextBlob


for word, freq in word_freq.items():
    specific_word = word
    blob = TextBlob(specific_word)
    
    # Get the sentiment polarity, which is a value between -1 and 1.
    sentiment_score = blob.sentiment.polarity
    data.append({'word' : specific_word, 'count' : freq, 'sentiment_score' : sentiment_score})

df1 = pd.DataFrame(data)
df1
```

##  â€¢ Find the count of these below words that are not suitable for airline reviews.
'delay', 'delayed', 'late', 'cancel', 'unprofessional'

```python
# Filter out common words
filtered_words = []
for word, freq in word_freq.most_common():
    if freq > 5 and word in ['delay', 'delayed', 'late', 'cancel', 'unprofessional']:
        filtered_words.append((word, freq))

# Sort the filtered words by frequency in descending order
most_common_complaints = sorted(filtered_words, key=lambda x: x[1], reverse=True)

# Display the most common complaints or issues
for word, freq in most_common_complaints:
    print(f"{word}: {freq}")
```
