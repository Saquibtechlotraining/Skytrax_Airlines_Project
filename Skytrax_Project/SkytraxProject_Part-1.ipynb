# ‣ Cleaning Reviews
Explanation of the below code:

Data Preprocessing - The below code is performing data preprocessing on the 'review' column of the DataFrame. It includes steps like removing punctuation, converting text to lowercase, splitting text into words, and lemmatizing the words.

Sentiment Analysis - While the code itself doesn't perform sentiment analysis, it prepares the text data for sentiment analysis by cleaning and transforming it.

New Column 'corpus'- The preprocessed text data is stored in a new column named 'corpus'. This column can be used for sentiment analysis or other text analysis tasks.

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords  # Use 'stopwords' instead of 'stopWords'
import re

lemma = WordNetLemmatizer()

# Assuming you have a DataFrame named 'df' and a column 'reviews_data'
reviews_data = df['review'].str.strip("? Trip Verified |")

# Create an empty list to collect cleaned data corpus
corpus = []

# Loop through each review, remove punctuations, convert to lowercase, lemmatize, and save to the corpus
for rev in reviews_data:
    rev = re.sub('[^a-zA-Z]', ' ', rev)
    rev = rev.lower()
    rev = rev.split()
    
    rev = [lemma.lemmatize(word) for word in rev if word not in set(stopwords.words('english'))]
    rev = " ".join(rev)
    corpus.append(rev)
```

```python
df['corpus'] = corpus
```
